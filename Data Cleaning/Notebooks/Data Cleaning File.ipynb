```python
import re

cols = ['Ref.']

df.loc[:, cols] = df.loc[:, cols].astype(str).applymap(
    lambda x: re.sub(r"[^a-zA-Z0-9]", "", x) 
)
```

```python
df.describe(include='all')
```

```python
df.columns = df.columns.str.replace('\xa0', ' ', regex=False)
print([col for col in df.columns])
```

```python
def clean_text(text):
    text = re.sub(r"\[\s*.*?\s*\]", "" , text)#remove anything inside [] including these.
    text = re.sub(r"[^a-zA-Z0-9]", "" , text)#remove non alphanumerics
    return text 
columns = ["Actual gross" , "Actual gross(in 2022 dollars)" , "Average gross"]
for column in columns:
    df[column] = df[column].apply(clean_text)

df
```

```python
def clean_peak(text):
    if not pd.isna(text):
        text =  re.sub(r"\[\s*.*?\s*\]", "" , text)#again remove any [ ] and anything inside it.
    return text
columns = ["Peak" , "All Time Peak"]
for column in columns:
    df[column] = df[column].apply(clean_peak)
df
```

```python
def clean_title(text):
    text =  re.sub(r"[^a-zA-Z ]", "" , text)#any character other than alphabets is cleared
    return text
df["Tour title"] = df["Tour title"].apply(clean_title)
```

```python
df["Ref."] = df["Ref."].replace("1516" , "16")
df["Ref."] = df["Ref."].replace("d" , "10")
```

```python
df
```

```python
numeric_cols = ["Actual gross" , "Actual gross(in 2022 dollars)" 
                , "Average gross" , "Shows" , "Ref." , "Peak" , "All Time Peak"]
for cols in numeric_cols :
    df[cols] = df[cols].astype("float")
```
